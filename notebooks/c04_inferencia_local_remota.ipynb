```python
# ============================================================
# COMPARAÇÃO: INFERÊNCIA LOCAL VS REMOTA (GROQ)
# ============================================================

import sys
import os
```

```python
# Adiciona a raiz do projeto ao sys.path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..')))
print("✅ Caminho adicionado:", sys.path[0])

# ============================================================
# 1. Importações
# ============================================================
from groq import Groq
import os as os_env
import time
import pandas as pd
```

```python
# ============================================================
# 2. Configuração da API Groq
# ============================================================
GROQ_API_KEY = os_env.getenv("GROQ_API_KEY")
if not GROQ_API_KEY:
    GROQ_API_KEY = input("Digite sua chave da API Groq: ")

client = Groq(api_key=GROQ_API_KEY)

# ============================================================
# 3. Modelo Local (Conceitual)
# ============================================================
print("\n" + "="*50)
print("INFERÊNCIA LOCAL (CONCEITUAL)")
print("="*50)

print("""
🔹 **Modelo Local** (exemplo: SmolLM2-1.7B-Instruct):
   - Requer download de ~3.4 GB
   - Executa em CPU/GPU local
   - Dados permanecem privados (não saem da máquina)
   - Latência: depende do hardware (pode ser lento)
   - Custo: zero (hardware próprio)

🔹 **Limitações observadas no projeto:**
   - Erros de memória (Memory Access Violation) em algumas máquinas
   - Respostas repetitivas e alucinações com modelos menores
   - Demora para carregar o modelo (até 2 minutos)
""")

# ============================================================
# 4. Inferência Remota (Groq API)
# ============================================================
print("\n" + "="*50)
print("INFERÊNCIA REMOTA (GROQ API)")
print("="*50)

print("\n⏱️ Teste de latência...")
start = time.time()

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "system", "content": "Você é um assistente de pesquisa."},
        {"role": "user", "content": "Explique os benefícios do RAG em 3 pontos."}
    ],
    max_tokens=150,
    temperature=0.3
)
end = time.time()

print(f"✅ Resposta recebida em {end - start:.2f} segundos.")
print("\n📝 Resposta do modelo:")
print(response.choices[0].message.content)

# ============================================================
# 5. Tabela Comparativa
# ============================================================
print("\n" + "="*50)
print("COMPARAÇÃO LOCAL vs REMOTA")
print("="*50)

comparacao = {
    "Critério": ["Privacidade", "Custo", "Latência", "Controle", "Configuração", "Qualidade"],
    "Local": ["Alta (dados não saem)", "Zero (hardware)", "Depende do hardware", "Total", "Complexa", "Variável"],
    "Remota (Groq)": ["Média (dados vão para API)", "Zero (plano gratuito)", "Muito baixa (0.5-2s)", "Parcial", "Simples", "Alta (70B)"]
}

df_comp = pd.DataFrame(comparacao)
print(df_comp.to_string(index=False))

# ============================================================
# 6. Conclusão
# ============================================================
print("\n" + "="*50)
print("CONCLUSÃO")
print("="*50)
print("""
📌 **Recomendação:**
   - A API Groq resolve os problemas de memória e latência.
   - A qualidade das respostas é superior, com modelos maiores (70B).
   - O plano gratuito é suficiente para prototipagem e uso moderado.
   - Para dados sensíveis, considere modelo local com quantização 4-bit.

🔹 **Vantagens da API:**
   - Sem necessidade de GPU local.
   - Atualização automática dos modelos.
   - Infraestrutura escalável.

🔸 **Desvantagens:**
   - Dependência de internet.
   - Dados enviados para servidores externos.
""")

print("\n✅ Notebook executado com sucesso!")
```
