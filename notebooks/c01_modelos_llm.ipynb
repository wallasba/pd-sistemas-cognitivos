```python
# Adiciona a raiz do projeto ao sys.path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..')))
print("✅ Caminho adicionado:", sys.path[0])

# ============================================================
# 1. Importações
# ============================================================
from groq import Groq
import os as os_env
```

```python
# ============================================================
# 2. Configuração da API
# ============================================================
# A chave pode vir do .env ou da variável de ambiente
GROQ_API_KEY = os_env.getenv("GROQ_API_KEY")
if not GROQ_API_KEY:
    print("⚠️ Chave da API Groq não encontrada. Defina GROQ_API_KEY no .env")
    # Em ambiente interativo, pode pedir ao usuário
    GROQ_API_KEY = input("Digite sua chave da API Groq: ")

client = Groq(api_key=GROQ_API_KEY)

# ============================================================
# 3. Teste de Geração Básica
# ============================================================
print("\n" + "="*50)
print("Teste 1: Geração básica")
print("="*50)

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "system", "content": "Você é um especialista em Inteligência Artificial."},
        {"role": "user", "content": "Explique o que é RAG (Retrieval-Augmented Generation) em uma frase."}
    ],
    max_tokens=100,
    temperature=0.3
)
print("Resposta:", response.choices[0].message.content)

# ============================================================
# 4. Comparação de Parâmetros (Temperatura)
# ============================================================
print("\n" + "="*50)
print("Teste 2: Comparação de temperatura")
print("="*50)

prompt = "Liste 3 aplicações da IA na saúde."

for temp in [0.0, 0.5, 1.0]:
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=80,
        temperature=temp
    )
    print(f"\nTemperatura {temp}:")
    print(response.choices[0].message.content)

# ============================================================
# 5. Modelos Disponíveis no Groq (Gratuitos)
# ============================================================
print("\n" + "="*50)
print("Modelos disponíveis no Groq (plano gratuito):")
print("="*50)
print("- llama-3.1-8b-instant (rápido, 8B parâmetros)")
print("- llama-3.3-70b-versatile (mais preciso, 70B parâmetros)")
print("- mixtral-8x7b-32768 (alto contexto, 32K tokens)")
print("- gemma2-9b-it (Google, 9B parâmetros)")

print("\n✅ Notebook executado com sucesso!")
```
