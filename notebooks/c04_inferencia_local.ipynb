```python
# %% [markdown]
# # c04 - Inferência Local vs Remota
# 
# Justificativa para execução local e análise de trade-offs.

# %%
import sys
sys.path.append('..')
from src.llm_inference import load_llm
import time

# %%
# Carrega um modelo pequeno (TinyLlama) e um médio (Phi-3) para comparar
modelos = [
    ("TinyLlama/TinyLlama-1.1B-Chat-v1.0", "1.1B"),
    ("microsoft/Phi-3-mini-4k-instruct", "3.8B")
]

for nome, tamanho in modelos:
    print(f"\nCarregando {nome} ({tamanho})...")
    inicio = time.time()
    try:
        pipe, _ = load_llm(nome)
        fim = time.time()
        print(f"  ⏱️ Tempo de carregamento: {fim - inicio:.2f}s")
        print(f"  💾 Memória estimada: ~{int(tamanho.replace('B',''))*2}GB")
    except Exception as e:
        print(f"  ❌ Erro: {e}")

# %%
# Análise Crítica:
# - Privacidade: dados não saem da máquina → ideal para pesquisador
# - Custo: zero (além do hardware)
# - Latência: depende do hardware, mas aceitável para uso interativo
# - Controle: total sobre o modelo e parâmetros

# %%
# Comparação com API cloud (discussão teórica):
# - OpenAI: alto custo, dados sensíveis expostos, latência baixa
# - Hugging Face Inference API: médio custo, dados expostos, dependência de internet
```
