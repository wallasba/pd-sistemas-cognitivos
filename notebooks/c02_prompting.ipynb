```python
# %% [markdown]
# # c02 - Prompt Engineering
# 
# Teste e comparação de diferentes técnicas de prompting.

# %%
import sys
sys.path.append('..')
from src.prompt_builder import build_prompt
from src.llm_inference import load_llm, generate_text

# %%
# Carrega o modelo (reutilizando o mesmo do notebook anterior)
pipe, tokenizer = load_llm("microsoft/Phi-3-mini-4k-instruct")

# %%
# Contexto de exemplo (um trecho de artigo)
contexto_exemplo = """
Deep learning has revolutionized medical imaging analysis. Convolutional neural networks (CNNs) achieve state-of-the-art 
performance in detecting tumors, retinal diseases, and pathological conditions. However, challenges remain regarding 
interpretability and data scarcity.
"""

pergunta = "Quais são os desafios do deep learning em imagens médicas?"

# %%
# Teste: Zero-Shot
messages_zs = build_prompt(pergunta, contexto_exemplo, "zero_shot")
prompt_zs = tokenizer.apply_chat_template(messages_zs, tokenize=False, add_generation_prompt=True)
resposta_zs = generate_text(pipe, tokenizer, prompt_zs)
print("=== ZERO-SHOT ===")
print(resposta_zs)

# %%
# Teste: Few-Shot
messages_fs = build_prompt(pergunta, contexto_exemplo, "few_shot")
prompt_fs = tokenizer.apply_chat_template(messages_fs, tokenize=False, add_generation_prompt=True)
resposta_fs = generate_text(pipe, tokenizer, prompt_fs)
print("=== FEW-SHOT ===")
print(resposta_fs)

# %%
# Teste: Chain-of-Thought
messages_cot = build_prompt(pergunta, contexto_exemplo, "cot")
prompt_cot = tokenizer.apply_chat_template(messages_cot, tokenize=False, add_generation_prompt=True)
resposta_cot = generate_text(pipe, tokenizer, prompt_cot)
print("=== CHAIN-OF-THOUGHT ===")
print(resposta_cot)

# %%
# Análise: Qual técnica produziu a resposta mais coerente e fundamentada?
# Documente a iteração e a escolha final.
```
