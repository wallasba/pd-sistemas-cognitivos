```python
# ============================================================
# COMPARAÇÃO DE TÉCNICAS DE PROMPT ENGINEERING
# ============================================================

import sys
import os
```

```python
# Adiciona a raiz do projeto ao sys.path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..')))
print("✅ Caminho adicionado:", sys.path[0])

# ============================================================
# 1. Importações
# ============================================================
from src.prompt_builder import build_prompt
from groq import Groq
import os as os_env
```

```python
# ============================================================
# 2. Configuração da API Groq
# ============================================================
GROQ_API_KEY = os_env.getenv("GROQ_API_KEY")
if not GROQ_API_KEY:
    GROQ_API_KEY = input("Digite sua chave da API Groq: ")

client = Groq(api_key=GROQ_API_KEY)

# ============================================================
# 3. Definir Contexto e Pergunta
# ============================================================
contexto = """
Deep learning tem revolucionado a análise de imagens médicas. Redes neurais convolucionais (CNNs) alcançam desempenho de ponta na detecção de tumores, doenças retinianas e condições patológicas. No entanto, desafios permanecem em relação à interpretabilidade e à escassez de dados.

A falta de transparência nos modelos dificulta a confiança dos médicos, e a necessidade de grandes conjuntos de dados rotulados limita a aplicação em doenças raras. Além disso, questões éticas como viés algorítmico e privacidade de dados são preocupações crescentes.
"""

pergunta = "Quais são os principais desafios do deep learning em imagens médicas?"

# ============================================================
# 4. Testar Técnicas de Prompt
# ============================================================
print("\n" + "="*50)
print("COMPARAÇÃO DE TÉCNICAS DE PROMPT")
print("="*50)

tecnicas = ["zero_shot", "few_shot", "cot"]

for tecnica in tecnicas:
    print(f"\n--- {tecnica.upper()} ---")
    
    # Constrói o prompt usando a função do projeto
    messages = build_prompt(pergunta, contexto, tecnica)
    
    # Para few-shot, adicionamos exemplos manualmente (simulação)
    if tecnica == "few_shot":
        messages = [
            {"role": "system", "content": "Você é um assistente de pesquisa."},
            {"role": "user", "content": "Exemplo: Quais são os desafios da IA na saúde? Resposta: Privacidade, viés e interpretabilidade."},
            {"role": "user", "content": f"Contexto:\n{contexto}\n\nPergunta: {pergunta}\nResposta:"}
        ]
    elif tecnica == "cot":
        messages = [
            {"role": "system", "content": "Você é um assistente de pesquisa."},
            {"role": "user", "content": f"Contexto:\n{contexto}\n\nPergunta: {pergunta}\n\nVamos pensar passo a passo:\n1. Identifique os desafios mencionados.\n2. Explique cada um.\n3. Resuma.\n\nResposta:"}
        ]
    else:  # zero_shot
        messages = [
            {"role": "system", "content": "Você é um assistente de pesquisa. Responda com base no contexto."},
            {"role": "user", "content": f"Contexto:\n{contexto}\n\nPergunta: {pergunta}\nResposta:"}
        ]
    
    # Envia para a API
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=messages,
        max_tokens=200,
        temperature=0.3
    )
    
    print("Resposta:\n", response.choices[0].message.content)
    print("-" * 40)

# ============================================================
# 5. Análise dos Resultados
# ============================================================
print("\n" + "="*50)
print("ANÁLISE DOS RESULTADOS")
print("="*50)
print("""
📌 Zero-Shot: Resposta direta, mas pode perder nuances.
📌 Few-Shot: Melhor para perguntas específicas, com exemplos.
📌 Chain-of-Thought: Mais detalhada, com raciocínio explícito.

💡 Recomendação: Para perguntas abertas, usar Chain-of-Thought.
   Para perguntas factuais, Zero-Shot é suficiente.
""")

print("\n✅ Notebook executado com sucesso!")
```
